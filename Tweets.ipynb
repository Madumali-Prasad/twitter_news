```python
import os
import pandas as pd
from datetime import date
```

```python
today = date.today()
end_date = today
```

```python
search_term = 'BBC News (World)', 'Ada Derana'
from_date = '2019-01-01'
```

```python
os.system(f"snscrape --since {from_date} twitter-search '{search_term} until:{end_date}' > result-tweets.csv")
if os.stat("result-tweets.csv").st_size == 0:
  counter = 0
else:
  df = pd.read_csv('result-tweets.csv', names=['tweets'])
  counter = df.size

print('Number Of Tweets : '+ str(counter))
```

```python
max_results = 500
```

```python
extracted_tweets = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-search '{search_term} until:{end_date}' > extracted-tweets.txt"
os.system(extracted_tweets)
if os.stat("extracted-tweets.csv").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('extracted-tweets.csv', names=['content'])
  for row in df['content'].iteritems():
    print(row)
```

```python
user_name = "BBC News (World)", "Ada Derana"
user_tweets = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-user '{user_name} until:{end_date}' > user-tweets.txt"
os.system(user_tweets)
if os.stat("user-tweets.txt").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('user-tweets.txt', names=['content'])
  for row in df['content'].iteritems():
    print(row)
```
